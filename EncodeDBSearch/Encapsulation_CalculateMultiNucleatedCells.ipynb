<h3> Read the RelateObject output from CellProfiler </h3>

```python
related_objects=pd.read_csv("MyExpt_RelateObjects.csv",sep=",")
```

```python
related_objects.head()
```

To relate the number of nuclei per cell we have to use <i>NUmber_Object_Number</i> and <i>Parent_NucleusObject</i>

```python
data_to_calculate_multinucleated_cells=related_objects.loc[:,['ImageNumber','ObjectNumber','Number_Object_Number','Parent_NucleusObject']]
```

First, for every image get the number of multinucleated objects

```python
dictionary_multi_nucleated_cells=dict()
```

```python
image_numbers=data_to_calculate_multinucleated_cells.ImageNumber.unique()
for x in image_numbers:
    tmp_data=data_to_calculate_multinucleated_cells.loc[data_to_calculate_multinucleated_cells['ImageNumber']==x,['Number_Object_Number','Parent_NucleusObject']]
    # Remove all the Parent_NucleusObject == 0 since they don't have an parent nucleus
    tmp_data=tmp_data.loc[tmp_data['Number_Object_Number']!=0,:]
    multi_nucleated_cell_count=tmp_data['Parent_NucleusObject'].duplicated().sum()
    dictionary_multi_nucleated_cells[x]=multi_nucleated_cell_count
```
